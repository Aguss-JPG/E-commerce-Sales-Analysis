# Business Analysis (ventas, devoluciones, datos inválidos)

- Resumen ejecutivo
Las ventas totales alcanzan 10.306.265,59 con devoluciones por 630.228,94, dejando ventas netas de 9.676.036,65. La tasa de devolución es 6,12%, un nivel moderado para e‑commerce. Los datos inválidos representan solo 0,7%, lo que indica buena calidad general del dataset y bajo riesgo de distorsión en KPIs.

- Ventas
El Reino Unido domina ampliamente el revenue con 8.846.129,26, concentrando la mayor parte del negocio. En un segundo nivel aparecen EIRE y Países Bajos, seguidos por Alemania y Francia. Esto sugiere una fuerte concentración geográfica, útil para priorizar campañas, logística y análisis de mix.
En productos, lideran “Manual”, “REGENCY CAKESTAND 3 TIER” y “WHITE HANGING HEART T-LIGHT HOLDER”. La presencia de conceptos como “POSTAGE” y “DOTCOM POSTAGE” en el top indica que costos de envío impactan el revenue reportado y deberían tratarse por separado si se busca analizar ventas puramente de producto.

- Devoluciones
El monto devuelto representa 6,12% de las ventas brutas. Es relevante pero no crítico; sugiere una oportunidad de optimización en ciertos productos o mercados. Para priorizar acciones, conviene cruzar tasa de devolución con volumen (producto‑país), enfocando en casos donde ambos sean altos.

- Datos inválidos
El 0,7% de registros inválidos es bajo. La limpieza aplicada probablemente no altera tendencias globales. Aun así, es útil monitorear si ese porcentaje crece en el tiempo o se concentra en ciertos países o productos.

```python
import pandas as pd
import numpy as np
```

```python
df_sales = pd.read_csv(r"C:\Users\ferra\OneDrive\Documentos\Data Projects\ecommerce-sales-analysis\data\proccesed\df_sales.csv")
df_returns = pd.read_csv(r"C:\Users\ferra\OneDrive\Documentos\Data Projects\ecommerce-sales-analysis\data\proccesed\df_return.csv")
df_invalid = pd.read_csv(r"C:\Users\ferra\OneDrive\Documentos\Data Projects\ecommerce-sales-analysis\data\proccesed\df_invalid.csv")
```

```python
# Asegurar Revenue
for d in [df_sales, df_returns]:
    d['Quantity'] = pd.to_numeric(d['Quantity'], errors='coerce')
    d['Price'] = pd.to_numeric(d['Price'], errors='coerce')
    d['Revenue'] = d['Quantity'] * d['Price']

# Totales generales
total_rows = len(df_sales) + len(df_returns) + len(df_invalid)
sales_revenue = df_sales['Revenue'].sum()
returns_amount = -df_returns['Revenue'].sum()
net_revenue = sales_revenue - returns_amount

invalid_rate = len(df_invalid) / total_rows * 100
return_rate = returns_amount / sales_revenue * 100 if sales_revenue else 0

print("Ventas totales:", round(sales_revenue,2))
print("Devoluciones totales:", round(returns_amount,2))
print("Ventas netas:", round(net_revenue,2))
print("Tasa de devolucion (%):", round(return_rate,2))
print("Datos invalidos (%):", round(invalid_rate,2))

# Top países por ventas
top_countries = (
    df_sales.groupby('Country', as_index=False)['Revenue']
    .sum().sort_values('Revenue', ascending=False).head(10)
)
print("\nTop paises ventas:")
print(top_countries.to_string(index=False))

# Top productos por ventas
top_products = (
    df_sales.groupby(['StockCode','Description'], as_index=False)['Revenue']
    .sum().sort_values('Revenue', ascending=False).head(10)
)
print("\nTop productos ventas:")
print(top_products.to_string(index=False))
```
